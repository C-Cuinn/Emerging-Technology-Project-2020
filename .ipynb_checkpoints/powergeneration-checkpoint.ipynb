# Power Generation Prediction

### Brief 
Create a web service that uses machine learning to make predictions based on the data set powerproduction.The goal is to
produce a model that accurately predicts wind turbine power output from wind speed values, as in the data set. You must then develop a web service that will respond with predicted power values based on speed values sent as HTTP requests. 

```python
%%time
# Magic Command that Displays Cell/Wall Times. Useful for Optimisation.

# Import packages.
import numpy as np
import pandas as pd
from flask import Flask
import scipy.stats as stats
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_squared_error
import seaborn as sns
import matplotlib.pyplot as plt
import pickle
# import tensorflow.keras as kr

%matplotlib inline

raw = pd.read_csv("./powerproduction.csv", delimiter = ',')# Import Raw Data.
```

### A Selection of Random Values from the Dataset to Verify that it Imported Correctly.

```python
raw.sample(15)# A Sample of Values from the Dataset.
```

### Null Values are Purged to Ensure they Do Not Effect the Results.

```python
%%time
ds = raw[(raw["speed"]==0) & (raw["power"]==0)]# Select Null Values.
not_null = raw[(raw["speed"]>0) & (raw["power"]>0)]# Keep all Values that ARE NOT Null
ds = ds.append(not_null)
print (ds)
print(ds.shape)
```

### An Overview of the Data.

```python
%%time
ds.describe(include = 'all')# A General Overview of Pertainent Statistics.
```

## Model Training Function



#### Now we put the speed and power values into arrays to use in training and testing.

```python
%%time
X = ds.iloc[:, 0].values.reshape(-1, 1) # Power Values Stored as Y.
Y = ds.iloc[:, 1].values.reshape(-1, 1) # Power Values Stored as Y.

A = np.array(raw[['speed']]) # Speed Values Stored as X.
B = np.array(raw[['power']]) # Power Values Stored as Y.

x = np.reshape(-1, 1)

fig, (ax1, ax2) = plt.subplots(1, 2,figsize=(15,10))

ax1.scatter(X,Y,color="blue",label='Data-point') # Plot a Graph with X(Wind Speed) vs Y(Power_Generation).
ax1.set_title('Null Values Purged')
ax1.set_xlabel('Wind_Speed')
ax1.set_ylabel('Power_Generation')
ax1.legend()

ax2.scatter(A,B,color="blue",label='Data-point') # Plot a Graph with X(Wind_Speed) vs Y(Power_Generation).
ax2.set_title('Raw Data')
ax2.set_xlabel('Wind_Speed')
ax2.legend()
```

####  we create an "X" and "Y" dataset for "X_train", "X_test" and "Y_train", "Y_test". Here we set 25% of the data as the test dataset and use the remaining for the training dataset. We create regressor then we fit "X_train" and "Y_train" into the regressor model.

# Regression : Linear & Polynomial

```python
%%time
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,train_size = 0.7,test_size = 0.3,random_state=10) # Creating Training & Testing Dataset.
regressor = LinearRegression() # Creating a Regressor.
regressor.fit(X_train,Y_train) # Fitting the Dataset into the Model.
```

#### Now we Display the Test Dataset (X_test and Y_test datasets) and draw the Regression Line.

```python
%%time
poly_reg = PolynomialFeatures(degree=4)
X_poly = poly_reg.fit_transform(X)
X_poly # prints X_poly
 
lin_reg2 = LinearRegression()
lin_reg2.fit(X_poly,Y)
```

```python
%%time
poly_reg = PolynomialFeatures(degree=4)
X_poly = poly_reg.fit_transform(X)
lin_reg2 = LinearRegression()
lin_reg2.fit(X_poly,Y)
```

```python
%%time
X_grid = np.arange(min(X),max(X),0.1)
X_grid = X_grid.reshape(len(X_grid),1) 

fig, (ax1, ax2) = plt.subplots(1, 2,figsize=(15,10))

ax1.plot(X_train,regressor.predict(X_train),color="blue",linewidth=1,label='Regression line')# Plot the Regression Line.
ax1.scatter(X_test,Y_test,color="green", label='Data-point')# Plot the Dataset on the Graph.
ax1.set_title('Linear (Test Set)')
ax1.set_xlabel('Wind_Speed')
ax1.set_ylabel('Power_Generation')
ax1.legend()

ax2.plot(X_grid, lin_reg2.predict(poly_reg.fit_transform(X_grid)),color='blue',label='Regression Line')# Plot the Regression Line.
ax2.scatter(X_test,Y_test,color="red",label='Data-point')
ax2.set_title('Polynomial (Test)')
ax2.set_xlabel('Wind_Speed')
ax2.legend()
```

Now we Display the Training Datasets (X_train and Y_train datasets) and draw the Regression Line.

```python
%%time
X_grid = np.arange(min(X),max(X),0.1)
X_grid = X_grid.reshape(len(X_grid),1) 

fig, (ax1, ax2) = plt.subplots(1, 2,figsize=(15,10))

ax1.plot(X_train,regressor.predict(X_train),color="blue",linewidth=1,label='Regression Line')# Plot the Regression Line.
ax1.scatter(X_train,Y_train,color="green",label='Data Point')# Plot the Dataset on the Graph.
ax1.set_title('Linear (training)')
ax1.set_xlabel('Wind_Speed')
ax1.set_ylabel('Power_Generation')
ax1.legend()

ax2.plot(X_grid, lin_reg2.predict(poly_reg.fit_transform(X_grid)),color='blue',label='Regression Line')# Plot the Regression Line.
ax2.scatter(X_train,Y_train,color="red",label='Data Point')# Plot the Dataset on the Graph.
ax2.set_title('Polynomial (Training)')
ax2.set_xlabel('Wind_Speed')
ax2.legend()
```

```python
%%time
Y_pred = regressor.predict(X_test)
print('R2 score: %.2f' % r2_score(Y_test,Y_pred))# R2 Score
```

#### The error is squared to reduce the impact of errors balancing each other out.

```python
%%time
print('Mean Error :',mean_squared_error(Y_test,Y_pred))# The Mean Error Squared
```

### The Function to Predict the Power_Generation using the Linear Model

```python
def power_gen_lin(power):# A Function to Predict the Power_Generation according to Wind_Speed using the Linear Model.
    result = regressor.predict(np.array(power).reshape(1, -1))
    filename = 'model_lin'
    outfile = open(filename,'wb')
    
    pickle.dump(result,outfile)
    outfile.close()
    
    return(result[0,0])
```

### The Function to Predict the Power_Generation using the Polynomial Model

```python
def power_gen_poly(power):# A Function to Predict the Power_Generation according to Wind_Speed using the Polynomial Model.
    result = lin_reg2.predict( poly_reg.fit_transform(np.array(power).reshape(1,-1)))
    filename = 'model_pol'
    outfile = open(filename,'wb')
    
    pickle.dump(result,outfile)
    outfile.close()
    
    return(result[0,0])
```

### Enter a Wind_Speed to Predict the Power_Generation using the Linear Model.

```python
%%time
wind_speed = float(input('Enter Wind_Speed : '))# Input a Wind_Speed Value.
print('Power_Generated : ',int(power_gen_lin(wind_speed)))# Print the Predicted Power_Generation.
```

#### The linear prediction is generally accurate with large margin of error as seen in the graph where the regression line only directly correlates to the data at 3 points while also beginning at -20 meaning a wind_speed value of 0->4 may produce negative results. 

### Enter a Wind_Speed to Predict the Power_Generation using the Polynomial Model.

```python
%%time
wind_speed = float(input('Enter Wind_Speed : '))# Input a Wind_Speed Value.
print('Power_Generated : ', int(power_gen_poly(wind_speed)))# Print the Predicted Power_Generation.
```

#### The polynomial model is substantially more accurate as can be seen the previous graph where the regression line correlates directly to the data model producing far more accurate predictions across the board.
